```python
import pandas as pd
import gc
```

```python
df = pd.read_csv('../ex04/fines.csv')
```

```python
%%timeit
def calculate_loop_append(df):
    results = []
    for i in range(0, len(df)):
        result = df.iloc[i]['Fines'] / df.iloc[i]['Refund'] * df.iloc[i]['Year']
        results.append(result)
    return results
df['Result'] = calculate_loop_append(df)
gc.collect()
```

```python
%%timeit
def calculate_with_iterrows(df):
    results = []
    for index, row in df.iterrows():
        result = row['Fines'] / row['Refund'] * row['Year']
        results.append(result)
    return results
df['Result'] = calculate_with_iterrows(df)
gc.collect()
```

```python
%%timeit
def calculate_apply(df):
    return df.apply(lambda row: row['Fines'] / row['Refund'] * row['Year'], axis=1) 
df['Result'] = calculate_apply(df)
gc.collect()
```

```python
%%timeit
def calculate_with_series(df):
    return df['Fines'] / df['Refund'] * df['Year']
df['Result'] = calculate_with_series(df)
gc.collect()
```

```python
%%timeit
def calculate_with_series(df):
    return df['Fines'].values / df['Refund'].values * df['Year'].values
df['Result'] = calculate_with_series(df)
gc.collect()
```
