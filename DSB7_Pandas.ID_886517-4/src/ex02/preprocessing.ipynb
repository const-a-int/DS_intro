```python
import pandas as pd

df = pd.read_csv(
    '../data/auto.csv', 
    sep=',',               
)
```

```python
df = pd.read_csv(
    '../data/auto.csv', index_col = 'ID'
)
```

```python
df.count()
```

```python
df = df.drop_duplicates(subset=['CarNumber', 'Make_n_model', 'Fines'], keep='last')
```

```python
df.count()
```

```python
df.isna().sum()
```

```python
df = df.dropna(thresh=500, axis=1)
```

```python
df.isna().sum()
```

```python
df.loc[:, 'Refund'] = df['Refund'].ffill()
```

```python
df.isna().sum()
```

```python
fines_mean = df['Fines'][df['Fines'] != 0].mean()
fines_mean
```

```python
df.loc[:,'Fines'] = df['Fines'].fillna(fines_mean)
refund_mean = df['Refund'].mean()
refund_mean
```

```python
df.isna().sum()
```

```python
df['Make'] = df['Make_n_model'].apply(lambda x: x.split()[0])
df['Model'] = df['Make_n_model'].apply(lambda x: x.split()[1] if len(x.split()) > 1 else None)
df = df.drop('Make_n_model', axis=1)
```

```python
df.isna().sum()
```
